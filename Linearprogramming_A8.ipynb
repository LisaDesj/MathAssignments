```python
# Import PuLP modeler functions
from pulp import * 
```

```python
M=100
N=100
```

```python
column=[j+1 for j in range(N)]
```

```python
row=[i+1 for i in range(M)]
```

```python
def a(i,j):
    if i==j: return 0
    if i==j-1: return 1
    if i<j-1: return -1
    if i==j+1: return -1
    if i>j+1: return 1
```

```python
a={i:{j:a(i,j) for j in column} for i in row}
```

```python
# Create the 'q1' variable to contain the problem data
q1 = LpProblem("Q1", LpMaximize)
```

```python
# A dictionary called 'decision_vars' is created to contain the referenced Variables
decision_vars = LpVariable.dicts("x", range(1,102,1)) 
```

```python
# The objective function is added to 'q1' first
q1 += decision_vars[101]
```

```python
# Adding constraints to 'q1'
for i in range(1,101,1):
    q1 += lpSum([-a[i][j]*decision_vars[j] for j in range(1,101,1)])+decision_vars[101] <= 0
    
q1 += lpSum(decision_vars[j] for j in range(1,101,1)) == 1

for j in range(1,102,1):
    q1 += decision_vars[j] >= 0
```

```python
# # The problem is solved using PuLP's choice of Solver
q1.solve()

# The status of the solution is printed to the screen
print("Status:", LpStatus[q1.status])
```

```python
# Each of the variables is printed with it's resolved optimum value
for a in q1.variables():
    print(a.name, "=", a.varValue)
```

```python
print("Optimal Value = ", value(q1.objective))
```

```python
# Printing Dual Variables
stop = '_C102'
for name, c in list(q1.constraints.items()):
    if name == stop:
        break
    print(name, "=", c.pi,)
```
