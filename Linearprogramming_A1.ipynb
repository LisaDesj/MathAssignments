```python
import pulp
```

```python
hw2_q1 = pulp.LpProblem("hw2_q1", pulp.LpMaximize)
```

```python
x1 = pulp.LpVariable('x1', upBound=4, lowBound=0, cat='Continuous')
x2 = pulp.LpVariable('x2', upBound=8, lowBound=0, cat='Continuous')
x3 = pulp.LpVariable('x3', upBound=22, lowBound=0, cat='Continuous')
x4 = pulp.LpVariable('x4', upBound=8, lowBound=0, cat='Continuous')
x5 = pulp.LpVariable('x5', upBound=13, lowBound=0, cat='Continuous')
x6 = pulp.LpVariable('x6', upBound=20, lowBound=0, cat='Continuous')
x7 = pulp.LpVariable('x7', upBound=3, lowBound=0, cat='Continuous')
x8 = pulp.LpVariable('x8', upBound=10, lowBound=0, cat='Continuous')
x9 = pulp.LpVariable('x9', upBound=18, lowBound=0, cat='Continuous')
```

```python
# Objective function
hw2_q1 += 300*x1 + 220*x2 + 100*x3 + 160*x4 + 130*x5 + 80*x6 + 360*x7 + 280*x8 + 140*x9, "Z"

# Constraints
hw2_q1 += x1 + x2 + x3 + x7 + x8 + x9 <= 30
hw2_q1 += x4 + x5 + x6 + x7 + x8 + x9 <= 30
```

```python
hw2_q1
```

```python
hw2_q1.solve()
pulp.LpStatus[hw2_q1.status]
```

```python
for variable in hw2_q1.variables():
    print ("{}={}".format(variable.name, variable.varValue))
```
