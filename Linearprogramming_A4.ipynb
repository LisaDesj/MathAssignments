```python
import pulp
```

```python
hw4_q2 = pulp.LpProblem("hw4_q2",pulp.LpMaximize)
```

```python
M=10
N=10
```

```python
column = [j+1 for j in range(N)]
row = [i+1 for i in range(M)]
```

```python
b = {i: (-2)**i for i in row}
```

```python
b
```

```python
a = {i: {j: (((-2)**(i+j))*((i**2)-(j**2))) for j in column} for i in row}
```

```python
a
```

```python
x = pulp.LpVariable.dicts("x",column,lowBound=0)
```

```python
for i in row:
    hw4_q2 += pulp.lpSum(a[i][j]*x[j] for j in column) <= b[i]
```

```python
hw4_q2
```

```python
hw4_q2.solve()
pulp.LpStatus[hw4_q2.status]
```

```python
for variable in hw4_q2.variables():
    print ("{}={}".format(variable.name, variable.varValue))
```
